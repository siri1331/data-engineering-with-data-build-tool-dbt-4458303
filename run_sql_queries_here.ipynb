```python
import duckdb
import pandas as pd
```

```python
sql_query = '''
show tables
'''

with duckdb.connect('data/nyc_parking_violations.db') as con:
    display(con.sql(sql_query).df())
```

```python
query_1 = '''
CREATE OR REPLACE TABLE parking_violation_codes AS
SELECT * 
FROM read_csv(
'data/dof_parking_violation_codes.csv',
auto_detect = True,
normalize_names = True
)
'''


query_2 = '''
CREATE OR REPLACE TABLE parking_violations_2023  AS
SELECT *
FROM read_csv(
'data/parking_violations_issued_fiscal_year_2023_sample.csv',
auto_detect = True,
normalize_names = True
)
'''

with duckdb.connect('data/nyc_parking_violations.db') as con:
  con.sql("""SHOW TABLES""")
  con.sql(query_1)
  con.sql(query_2)
```

```python
sql_query = '''
select * FROM parking_violation_codes limit 3
'''

with duckdb.connect('data/nyc_parking_violations.db') as con:
    display(con.sql(sql_query).df())
```

```python
sql_query = '''
select * FROM parking_violations_2023 limit 3
'''

with duckdb.connect('data/nyc_parking_violations.db') as con:
    display(con.sql(sql_query).df())
```
